```python
import numpy as np
import matplotlib.pyplot as plt
import os, sys
import spiceypy as spice
from mpl_toolkits.mplot3d import Axes3D
```

```python
#Screw trying to manage the SPICE, load all of the kernels.
#Ideally, one would be able to pull the specific kernels required for the time periods of interest.
#This is waiting on a SPICE server.
spice.kclear()
spice.tkvrsn("TOOLKIT")
spice.furnsh("/tmp/junokerns.krn")
```

```python
ptimes4 = []
ptimes3 = []
#Observation times taken from the ORBIT_04 images.
obs_f4 = '/Volumes/Work/JunoCam/examples/images/data/juno/JNOJNC_0003/DATA/EDR/JUPITER/ORBIT_04/observation_times'
with open(obs_f4) as f:
    for line in f:
        net = spice.utc2et(line.strip())
        ptimes4.append(net)
#Observation times taken from the ORBIT_03 images.
#This is here as an example, but I'm not plotting these observations or any part of the orbit.
#obs_f3 = '/Volumes/Work/JunoCam/examples/images/data/juno/JNOJNC_0003/DATA/EDR/JUPITER/ORBIT_03/obs_times'
#with open(obs_f3) as f:
#    for line in f:
#        net = spice.utc2et(line.strip())
#        ptimes3.append(net)

#Not plotting orbit 3, so we don't need to do the calculations.
#startT3 = spice.utc2et("2016-12-10T17:15:14.358")
#endT3 = spice.utc2et("2017-01-27T08:15:15.528")
startT4 = spice.utc2et("2017-01-27T08:15:15.528")
endT4 = spice.utc2et("2017-02-08T05:45:26.848")
#Step decides the precision of the position calculation.
step = 10000
#times3 = [x*(endT3-startT3)/step + startT3 for x in range(step)]
times4 = [x*(endT4-startT4)/step + startT4 for x in range(step)]
```

```python
#The next step is to plot the look vectors from a given instrument.
```

```python
plot_ref = [ 'J2000', \
            'IAU_JUPITER']

for ref in range(len(plot_ref)):
    #pos3, lightTimes = spice.spkpos('Juno', times3, plot_ref[ref], 'NONE', 'JUPITER')
    #pos3=np.array([np.array(ni) for ni in pos3])
    pos4, lightTimes = spice.spkpos('Juno', times4, plot_ref[ref], 'NONE', 'JUPITER')
    pos4=np.array([np.array(ni) for ni in pos4])
    p_positions4, p_lights = spice.spkpos('Juno', ptimes4, plot_ref[ref], 'None', 'JUPITER' )
    p_positions4 = np.array([np.array(ni) for ni in p_positions4])

    fig = plt.figure(figsize=(30, 30))
    ax = Axes3D(fig)
    #ax.plot(pos3[:,0], pos3[:,1], pos3[:,2], color="purple")
    ax.plot(pos4[:,0], pos4[:,1], pos4[:,2], color="orange")
    ax.plot(p_positions4[:,0], p_positions4[:,1], p_positions4[:,2], 'bx', markersize=10)
    
    plt.show()
    
```
